```python
import pandas as pd
import numpy as np
from pykalman import KalmanFilter
import matplotlib.pyplot as plt

# c d liu
df = pd.read_csv(r"c:\timeseries\train.csv")
df1 = df[df["Publication_Day"] == "Thursday"]
df1 = df1.drop("id", axis=1)
df2 = df1[df1["Publication_Time"] == "Morning"]
observations = df2['Listening_Time_minutes'].values

# Cu hnh Kalman Filter cho m hnh t hi quy (AR(1))
phi = 0.9  #
kf = KalmanFilter(
    transition_matrices=[[phi]],
    observation_matrices=[[1]],
    initial_state_mean=[observations[0]],
    initial_state_covariance=[[1]],
    observation_covariance=1,
    transition_covariance=[[0.1]]
)

# p dng Kalman Filter
state_means, _ = kf.filter(observations)

# V biu 
plt.figure(figsize=(14, 5))
plt.plot(observations, label="Original")
plt.plot(state_means.flatten(), label="Kalman Autoregressive")
plt.title("Kalman Filter with Autoregressive Model on Thursday Morning Listening Time")
plt.legend()
plt.savefig("kalman_autoregressive.png")
```

```python
import pandas as pd
import numpy as np
from pykalman import KalmanFilter
import matplotlib.pyplot as plt

# c d liu
df = pd.read_csv(r"c:\timeseries\train.csv")
df1 = df[df["Publication_Day"] == "Thursday"]
df1 = df1.drop("id", axis=1)
df2 = df1[df1["Publication_Time"] == "Morning"]
observations = df2['Listening_Time_minutes'].values

# Cu hnh Kalman Filter cho m hnh xu hng tuyn tnh
kf = KalmanFilter(
    transition_matrices=[[1, 1], [0, 1]],  # Ma trn chuyn tip: [level, slope]
    observation_matrices=[[1, 0]],         # Ch quan st level, khng quan st slope
    initial_state_mean=[observations[0], 0],  # Gi tr ban u: [level, slope]
    initial_state_covariance=[[1, 0], [0, 1]],  #  khng chc chn ban u
    observation_covariance=1,              # Nhiu php o
    transition_covariance=[[0.01, 0], [0, 0.01]]  # Nhiu chuyn tip
)

# p dng Kalman Filter
state_means, _ = kf.filter(observations)

# V biu 
plt.figure(figsize=(14, 5))
plt.plot(observations, label="Original")
plt.plot(state_means[:, 0], label="Kalman Linear Trend")
plt.title("Kalman Filter with Linear Trend on Thursday Morning Listening Time")
plt.legend()
plt.savefig("kalman_linear_trend.png")
```

```python
import pandas as pd
import numpy as np
from pykalman import KalmanFilter
import matplotlib.pyplot as plt

# c d liu
df = pd.read_csv(r"c:\timeseries\train.csv")
df1 = df[df["Publication_Day"] == "Thursday"]
df1 = df1.drop("id", axis=1)
df2 = df1[df1["Publication_Time"] == "Morning"]
observations = df2['Listening_Time_minutes'].values

# 1 thng
seasonal_period = 4
n_states = seasonal_period + 1  # Trng thi: level + cc thnh phn ma v

# Cu hnh Kalman Filter cho m hnh ma v
transition_matrix = np.zeros((n_states, n_states))
transition_matrix[0, 0] = 1  # Level khng i
for i in range(seasonal_period):
    transition_matrix[1 + i, 1 + (i % seasonal_period)] = 1  # Chu k ma v
observation_matrix = np.array([[1] + [0] * seasonal_period])  # Ch quan st level
initial_state_mean = [observations[0]] + [0] * seasonal_period  # Gi tr ban u
initial_state_covariance = np.eye(n_states)  #  khng chc chn ban u
observation_covariance = 1  # Nhiu php o
transition_covariance = np.eye(n_states) * 0.01  # Nhiu chuyn tip

kf = KalmanFilter(
    transition_matrices=transition_matrix,
    observation_matrices=observation_matrix,
    initial_state_mean=initial_state_mean,
    initial_state_covariance=initial_state_covariance,
    observation_covariance=observation_covariance,
    transition_covariance=transition_covariance
)

# p dng Kalman Filter
state_means, _ = kf.filter(observations)

# V biu 
plt.figure(figsize=(14, 5))
plt.plot(observations, label="Original")
plt.plot(state_means[:, 0], label="Kalman Seasonal")
plt.title("Kalman Filter with Seasonality on Thursday Morning Listening Time")
plt.legend()
plt.savefig("kalman_seasonal.png")
```
